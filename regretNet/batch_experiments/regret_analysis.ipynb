```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
long_dp_off = pd.read_csv('exp_5x10_long_dp_off.csv')
long_dp_zero = long_dp_off.replace({'Noise': None}, 0).replace({'Clip': None}, 0)
```

```python
sns.set(rc={'figure.figsize':(15,15)})
sns.lineplot(x='Iter', y='Regret', data=long_dp_off)
sns.lineplot(x='Iter', y='Revenue', ax=plt.twinx(), data=long_dp_off, color='red')
```

```python
noise_1_clip_1 = pd.read_csv('additive_5x10_uniform_noise_1_clip_1.csv')
noise_1_3_clip_1 = pd.read_csv('additive_5x10_uniform_noise_1.3_clip_1.csv')
noise_1_5_clip_1 = pd.read_csv('additive_5x10_uniform_noise_1.5_clip_1.csv')
noise_1_8_clip_1 = pd.read_csv('additive_5x10_uniform_noise_1.8_clip_1.csv')
noise_2_clip_1 = pd.read_csv('additive_5x10_uniform_noise_2_clip_1.csv')

noise_1_clip_10 = pd.read_csv('additive_5x10_uniform_noise_1_clip_10.csv')
noise_1_3_clip_10 = pd.read_csv('additive_5x10_uniform_noise_1.3_clip_10.csv')
noise_1_5_clip_10 = pd.read_csv('additive_5x10_uniform_noise_1.5_clip_10.csv')
noise_2_clip_10 = pd.read_csv('additive_5x10_uniform_noise_2_clip_10.csv')

noise_1_clip_1_long = pd.read_csv('additive_5x10_uniform_noise_1_clip_1_long.csv')
noise_1_8_clip_1_long = pd.read_csv('additive_5x10_uniform_noise_1.8_clip_1_long.csv')
noise_2_clip_1_long = pd.read_csv('additive_5x10_uniform_noise_2_clip_1_long.csv')

clip_1 = [noise_1_clip_1, noise_1_3_clip_1, noise_1_5_clip_1, noise_1_8_clip_1, noise_2_clip_1]
clip_10 = [noise_1_clip_10, noise_1_3_clip_10, noise_1_5_clip_10, noise_2_clip_10]

all_noise = pd.concat([long_dp_zero, *clip_1, *clip_10]).query('Iter<=100000')
all_noise_long = pd.concat([long_dp_zero, noise_1_clip_1_long, noise_1_8_clip_1_long, noise_2_clip_1_long])
```

```python
v5 = sns.FacetGrid(all_noise, col='Noise', row='Clip')
v5 = v5.map(plt.plot, 'Iter', 'Revenue')
```
