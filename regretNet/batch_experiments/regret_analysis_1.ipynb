```python
import pandas as pd
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
low_noise = pd.read_csv('exp_5x10_short_10_dp_low_noise_1.csv')
dp_off = pd.read_csv('exp_5x10_short_10_dp_off.csv')
long_dp_off = pd.read_csv('exp_5x10_long_dp_off_2.csv')
iter4k_low_noise = low_noise.query('Iter==4000')
iter4k_dp_off = dp_off.query('Iter==4000')
```

```python
ttest_ind(low_noise['Revenue'], dp_off['Revenue'])
```

```python
rev_mean_short = iter4k_low_noise['Revenue'].mean() 
reg_mean_short = iter4k_low_noise['Regret'].mean()
rev_std_short = iter4k_low_noise['Revenue'].std()
reg_std_short = iter4k_low_noise['Regret'].std()
last_rev_long = long_dp_off.iloc[-1]['Revenue']
last_reg_long = long_dp_off.iloc[-1]['Regret']
```

```python
(rev_mean_short - last_rev_long)/rev_std_short
```

```python
(reg_mean_short - last_reg_long)/reg_std_short
```

```python
sns.set(rc={'figure.figsize':(15,15)})
sns.lineplot(x='Iter', y='Regret', data=long_dp_off)
```
